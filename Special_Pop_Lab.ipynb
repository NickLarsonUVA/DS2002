**No celebrity planes in this API data set had to try another**

```python
import requests

def get_plane_position(icao24):
    url = "https://opensky-network.org/api/states/all"

    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an error for bad status codes
        data = response.json()

        for state in data.get("states", []):
            if state[0].lower() == icao24.lower():
                return {
                    "icao24": state[0],
                    "callsign": state[1].strip() if state[1] else "N/A",
                    "origin_country": state[2],
                    "time_position": state[3],
                    "last_contact": state[4],
                    "longitude": state[5],
                    "latitude": state[6],
                    "geo_altitude": state[7],
                    "on_ground": state[8],
                    "velocity": state[9],
                    "true_track": state[10],
                    "vertical_rate": state[11],
                    "baro_altitude": state[13],
                    "squawk": state[14] if state[14] else "N/A",
                    "spi": state[15],
                    "position_source": state[16],
                }

        return None  # No matching ICAO24 found

    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        return None

if __name__ == "__main__":
    icao24_code = input("Enter the ICAO24 code of the aircraft: ")
    plane_data = get_plane_position(icao24_code)

    if plane_data:
        print("\nAircraft Position Data:")
        for key, value in plane_data.items():
            print(f"{key}: {value}")
    else:
        print("No data found for the given ICAO24 code.")
```
